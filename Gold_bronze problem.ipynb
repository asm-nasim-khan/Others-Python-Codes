```python
import numpy as np
listCountry = {
    'Dhaka':[2,3,4],
    'Ctg':[4,4,3],
    'Gazipur':[2,4,4],
    'Raj':[2,3,4]
}
country=['Dhaka','Ctg','Gazipur','Raj']
a = np.zeros((4,3), dtype = np.int32)
#a[7] = 1
#print(a[7,2])
#print(a)
seq = []
for i in range(4):
    x = listCountry[country[i]]
    a[i,0] = int(x[0])
    a[i,1] = int(x[1])
    a[i,2] =int(x[2])
    
for i in range(4):
    if a[i,0]!= 0:
        maxn = a[i,0]
        maxLoc = 0
        maxcont= 0
        for j in range(1,4):
            if a[j,0] > maxn:
                maxn = a[j,0]
                maxLoc = j
                maxCount = 1
                a[j] = 0
            elif a[j,0] == maxn:
                maxCount += 1
        if maxCount == 1:
            seq.append(maxLoc)
            
for i in range(4):
    if a[i,1]!= 0:
        maxn = a[i,1]
        maxLoc = 0
        maxcont= 0
        for j in range(1,4):
            if a[j,1] > maxn:
                maxn = a[j,1]
                maxLoc = j
                maxCount = 1
                a[j] = 0
            elif a[j,1] == maxn:
                maxCount += 1
        if maxCount == 1:
            seq.append(maxLoc)
for i in range(4):
    if a[i,2]!= 0:
        maxn = a[i,2]
        maxLoc = 0
        maxcont= 0
        for j in range(1,4):
            if a[j,2] > maxn:
                maxn = a[j,2]
                maxLoc = j
                maxCount = 1
                a[j] = 0
            elif a[j,2] == maxn:
                maxCount += 1
        if maxCount == 1:
            seq.append(maxLoc)
alpha = []
for k in range(4):
    if a[k,0] != 0:
        alpha.append(country[k])
sortCountry = sorted(alpha)

for p in range(len(seq)): # printing Gold , Silver , Bronze part
    print(country[seq[p]]," : ",listCountry[country[seq[p]]]) 
for item in sortCountry: # printing Aalphabetically 
    print(item," : ",listCountry[item]) 
```
